```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.columns
```

```python
test_df.columns
```

```python
#Separate target feature for training data
x_train = train_df.drop("loan_status", axis=1)
y_train = train_df["loan_status"]

#Convert categorical data to numeric 
y_train = LabelEncoder().fit_transform(train_df["loan_status"])


x_train_dummy = pd.get_dummies(x_train)
```

```python
x_train_dummy
```

```python
#Separate target feature for testing data
x_test = test_df.drop("loan_status", axis=1)
y_test = test_df["loan_status"]

#Convert categorical data to numeric
y_test = LabelEncoder().fit_transform(test_df["loan_status"])

x_test_dummy = pd.get_dummies(x_test)
```

```python
#locate missing dummy variables
for i in x_train_dummy.columns:
    if i not in x_test_dummy.columns:
        print(i)
```

```python
# add missing dummy variables to testing set
x_test_dummy['debt_settlement_flag_Y'] = 0
x_test_dummy.head()
```

Consider the models:

After reading some articles about different models I believe Random Forests will perform better given the amount of features this data set has and it's ability to weight certain features as more important.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
clf = LogisticRegression().fit(x_train_dummy, y_train)

print(f'training score: {clf.score(x_train_dummy, y_train)}')
print(f'testing score: {clf.score(x_test_dummy, y_test)}')
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier().fit(x_train_dummy, y_train)
print(f'training score: {clf.score(x_train_dummy, y_train)}')
print(f'testing score: {clf.score(x_test_dummy, y_test)}')
```

It appears that the Random Forest model performed better than the Logistic Regression model.

```python
# Scale the data
scaler = StandardScaler().fit(x_train_dummy)
x_train_scaled = scaler.transform(x_train_dummy)
x_test_scaled = scaler.transform(x_test_dummy)
```

Prediction:
I predict both models will preform better with scaled data. 

```python
# Train the Logistic Regression model on the scaled data and print the model score
clf = LogisticRegression().fit(x_train_scaled, y_train)
print(f'training score: {clf.score(x_train_scaled, y_train)}')
print(f'testing score: {clf.score(x_test_scaled, y_test)}')
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier().fit(x_train_scaled, y_train)
print(f'training score: {clf.score(x_train_scaled, y_train)}')
print(f'testing score: {clf.score(x_test_scaled, y_test)}')
```

Conclusion:
After scaling the data the Linear Regression model had an improvement in score as predicted. The Random Forest model had a slight increase in score after scaling which is also what was predicted.
